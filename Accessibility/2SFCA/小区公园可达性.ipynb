```python
import json
import geopandas as gpd
import warnings
import pandas as pd
```

```python
#将数据转换为geopandas所支持的格式
#小区res
df = pd.read_excel("C:/Users/Administrator/.anaconda/南山区居住小区poi.xlsx")
res = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lng, df.lat)) #指定经纬度坐标所在列
res.crs = 'EPSG:4326' #指定坐标系 epsg4326为WGS1984
res.plot()
#公园park
df2 = pd.read_excel("C:/Users/Administrator/.anaconda/南山区公园poi.xlsx")
park = gpd.GeoDataFrame(df2, geometry=gpd.points_from_xy(df2.lng, df2.lat)) #指定经纬度坐标所在列
park.crs = 'EPSG:4326' #指定坐标系 epsg4326为WGS1984
park.plot()
```

```python
#转换坐标系 转换为可计算的投影坐标系
res = res.to_crs(epsg=3857)
park = park.to_crs(epsg=3857)
res
```

```python
#第一步计算
#对公园进行缓冲
park_s1 = park.copy() #复制
park_s1['geometry'] = park_s1['geometry'].buffer(2500) #缓冲2500米
park_s1['geometry'].plot()
park_s1
```

```python
park_s1 = gpd.sjoin(park_s1,res,op='contains') #建立空间连接，将在缓冲区内的小区连接到公园中
park_s1
```

```python
#计算所有OD的直线距离
from math import radians, cos, sin, asin, sqrt
def haversine(lon1,lat1,lon2,lat2): #经纬度;十进制度数
    #将十进制转化为弧度
    lon1,lat1,lon2,lat2 = map(radians,[lon1,lat1,lon2,lat2])
    #haversine公式
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371
    return c * r *1000
```

```python
#求得公园与小区之间的直线距离
park_s1['length'] = park_s1[['lng_left','lat_left','lng_right','lat_right']].\
apply(lambda x:haversine(x[0],x[1],x[2],x[3]),axis=1)
park_s1.head()
```

```python
#计算公园供需比Rj (缓冲半径为2.5km时，公园内的小区居住人数)
import numpy as np
def get_G(x):
    if x <= 2500:
        G = (np.exp(-0.5 * (x/2500)**2) - np.exp(-0.5)) / (1-np.exp(-0.5))
        return G
    else:
        return 0
def get_Rj(x):
    x = x.reset_index()
    sj = x['area'][0]
    dt = 0
    for i in range(len(x)):
        vl = x['pop'][i] * get_G(x['length'][i])
        dt += vl
    return sj/dt
```

```python
park_vk = park_s1.groupby(by='名称_left').apply(get_Rj).reset_index()
park_vk.head()
```

```python
park_vk.columns = ['名称','vl_R']
park_concat = pd.merge(park,park_vk,on='名称')
park_concat.head(10)
```

```python
#第二步计算
res_s2 = res.copy()
res_s2['geometry'] = res_s2['geometry'].buffer(2500)
res_s2 = gpd.sjoin(res_s2,park_concat,op='intersects')
res_s2['length'] = res_s2[['lng_left','lat_left','lng_right','lat_right']].\
apply(lambda x:haversine(x[0],x[1],x[2],x[3]),axis=1)
res_s2.head(1)
```

```python
#计算小区的公园可达性（小区半径为2.5km的范围内，公园供需比之和）
def get_G(x):
    if x <= 2500:
        G = (np.exp(-0.5 * (x/2500)**2) - np.exp(-0.5)) / (1-np.exp(-0.5))
        return G
    else:
        return 0
def get_A(x):
    x = x.reset_index()
    dt = 0
    for i in range(len(x)):
        vl = x['vl_R'][i] * get_G(x['length'][i])
        dt += vl
    return dt
```

```python
res_vl = res_s2.groupby(by='名称_left').apply(get_A).reset_index()
res_vl.columns=['名称','vl_R']
res_vl.head()
```

```python
dt = res.merge(res_vl,on='名称')
dt.head(10)
```

```python
#导出为shp文件
dt.to_file('小区公园可达性.shp',encoding = 'gb18030')
```
