```python
a = linkdf.head(1)
```

```python
url = a["链接"][0]
```

```python
a = requests.get(url,headers=headers)
html = a.text
soup = BeautifulSoup(html, 'lxml')
```

```python
for i in soup.find_all("ul",attrs={"id":"charu"})[0].find_all("li"):
    name = i.find_all("a")[0].text
    pinyin = ""
    for i in lazy_pinyin("name"):
        pinyin = pinyin + i[0]
    print(pinyin)
    urls = url + pinyin + "/jggk/"
    print(urls)
```
