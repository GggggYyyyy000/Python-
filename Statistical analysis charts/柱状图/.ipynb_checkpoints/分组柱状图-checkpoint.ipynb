```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei'] # 指定默认字体：解决plot不能显示中文问题
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

```python
data2 = pd.read_excel("/Users/creative/Documents/python/Statistical analysis charts/曲线图/描述性统计02.xlsx",sheet_name="公园数量")
lab = data2["行政区"].unique().tolist()
```

```python
data2["类型"].unique().tolist()
```

```python
group2 = data2.groupby("行政区")
```

```python
cs_park = []
sq_park = []
zr_park = []
#zh_park = []
for i in lab:
    cs_park.append(group2.get_group(i)["数量"].tolist()[0])
    sq_park.append(group2.get_group(i)["数量"].tolist()[1])
    zr_park.append(group2.get_group(i)["数量"].tolist()[2])
    #zh_park.append(group2.get_group(i)["数量"].tolist()[3])
```

```python
fig = plt.figure(figsize=(15,5),dpi=300)
ax = fig.add_subplot(1,1,1)

x = np.arange(len(lab))  # the label locations
width = 0.2  # the width of the bars

ax.bar(x - width/2, sq_park, width, label='社区公园',color = "#a2b881",edgecolor='black',linewidth=2)
#ax.bar(x + width/2, zl_park, width, label='专类公园',color = "#6d844d",edgecolor='black',linewidth=3)
ax.bar(x + width/2, zr_park, width, label='自然公园',color = "#45673b",edgecolor='black',linewidth=2)
ax.bar(x + width*1.5, cs_park, width, label='城市公园',color = "#000000",edgecolor='black',linewidth=2)

# Add some text for labels, title and custom x-axis tick labels, etc.

ax=plt.gca();#获得坐标轴的句柄
ax.spines['bottom'].set_linewidth(2);###设置底部坐标轴的粗细
ax.spines['left'].set_linewidth(2);####设置左边坐标轴的粗细
ax.spines['top'].set_linewidth(2);###设置底部坐标轴的粗细
ax.spines['right'].set_linewidth(2);####设置左边坐标轴的粗细
font2 = {'weight' : 'bold','size' : 23,}

plt.tick_params(labelsize=23)
labels = ax.get_xticklabels() + ax.get_yticklabels()
ax.set_ylabel('数量',font2)
ax.set_title('研究公园分区分类型比较',font2)
ax.set_xticks(x)
ax.set_xticklabels(lab,font2)
ax.legend(loc = "upper center",fontsize='xx-large',ncol=4)
plt.grid(linestyle= "-.")
fig.tight_layout()

plt.savefig("分类型比较图.jpeg",bbox_inches = 'tight')
```
