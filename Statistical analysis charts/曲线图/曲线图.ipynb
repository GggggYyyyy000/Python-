```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei'] # 指定默认字体：解决plot不能显示中文问题
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

```python
df = pd.read_excel("data.xlsx")
```

```python
df1 = df.loc[df["消费"] != 0 ].sort_values('消费',ascending=False)
df2 = df.loc[df["供给"] != 0 ].sort_values('供给',ascending=False)
```

```python
def arrays(df,cols):
    arrays = df.loc[:,[cols]][cols].values.tolist()
    #arrays = np.array(arrays)
    return arrays
```

```python
fig, axs = plt.subplots(2, 1, sharex=True)
fig.subplots_adjust(hspace=0)

axs[0].plot(arrays(df2,"city"), arrays(df2,"供给") ,label="生产")
axs[0].grid(alpha=0.4)
axs[0].legend()

axs[1].plot(arrays(df1,"city"), arrays(df1,"消费"), label="消费",color="r")
axs[1].grid(alpha=0.4)
axs[1].legend()

plt.xticks([])
plt.show()
```
