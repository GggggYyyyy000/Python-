```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei'] # 指定默认字体：解决plot不能显示中文问题
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

```python
tips = pd.read_excel("/Users/creative/Desktop/RL2015VSRL2021.xlsx")
```

```python
tips.corr()
```

```python
sns.set_style("white")
```

```python
sns.palplot(sns.diverging_palette(10, 220, sep=80, n=7))
```

```python
#sns.pairplot(tips, palette="Paired") 

g = sns.PairGrid(tips)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
sns.heatmap(tips.corr())
```

```python
# 分层相关性热力图
sns.clustermap(tips.corr())
```

```python
g = sns.PairGrid(tips)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.jointplot(x = 'intersection_density', y = 'NW_gcj', data = tips)
```

```python
sns.jointplot(x = 'sumRL2015', y = 'sumRL2021', data = tips ,kind = 'hex')
```

```python
sns.jointplot(x = 'sumRL2015', y = 'sumRL2021', data = tips ,kind = 'reg',color="k", line_kws={"color":"r","alpha":0.3,"lw":3})
```

```python
sns.jointplot(x = 'intersection_density', y = 'NW_gcj', data = tips ,kind = 'kde')
```
